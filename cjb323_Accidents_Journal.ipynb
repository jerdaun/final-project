Introduction: The dataset I am working with is accidental overdoses in the allegheny county area. In this case, the best neighboorhood will be th eone who has the lowest ammount of overdoses.

Dataset: https://data.wprdc.org/dataset/allegheny-county-fatal-accidental-overdoses/resource/1c59b26a-1684-4bfb-92f7-205b947530cf

```python
import pandas as pd

csvPath = "Overdoses to neighboorhood no PO box.csv"
table = pd.read_csv(csvPath, dtype='object')
df = pd.DataFrame(table)
sortColumn = 'AMOUNT'
df[sortColumn] = pd.to_numeric(df[sortColumn], errors="coerce")
df = df.sort_values(by=sortColumn)
scoreArray = []
i = 1
while i < 40:
    scoreArray.insert(0,0.25641025641*i)
    i+=1
df.insert(loc=1, column='SCORE', value=scoreArray)
df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

csvPath = "Overdoses to neighboorhood no PO box.csv"
table = pd.read_csv(csvPath, dtype='object')
df = pd.DataFrame(table)
sortColumn = 'AMOUNT'
df[sortColumn] = pd.to_numeric(df[sortColumn], errors="coerce")
df = df.sort_values(by=sortColumn)
plt.rcParams['font.size'] = 6
plt.bar(df["NEIGHBOORHOOD"],df["AMOUNT"])
plt.xticks(rotation=45, ha='right')
plt.show()
```

I first imported the data of every zipcode, what neighboorhood or neighboorhoods they encompass, and how many accidental overdoses there are in that region. I then sort them by the ammount with pandas, which gives the output table seen above. This dataset excludes the zip codes which point to PO boxes, for obvious reasons. The data is then also represented in a bargraph using numpy. This initially was impossible to read, as the long label names just overlapped into a black blob. So I also rotated the labels and made their font smaller so that the data is still readable.
The score collum was implemented so that all three metrics could be easier merged to produce a final answer. It scales linearly such that the best neighboorhood is 10.
