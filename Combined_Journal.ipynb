Introduction:
    We are trying to find the best neighborhood in Pittsburgh by using 3 different metrics that are mostly independent from eachother to find a neighborhood that is the 'best' in a wide spread of metrics.

Metric:
    The metric will be the average of the rankings of the neighborhoods in each metric. The features we a measuring are car crashes, fatal overdoses, and walkability/sidewalk to street ratio. 

Datasets:
    Paving Schedu: https://data.wprdc.org/dataset/paving-schedule
    Fatal Overdoses:
    Walkability: https://data.wprdc.org/dataset/sidewalk-to-street-walkability-ratio

The Best Neighborhood:
    

```python
df1 = pd.read_csv("ratio_scores.csv")
df2 = pd.read_csv("score_set_2.csv")
df3 = pd.read_csv("score_set_3.csv")

# Merge all on the neighborhood field
merged = df1.merge(df2, on="neighborhood").merge(df3, on="neighborhood")

# Simple average (all equal weight)
merged["final_score"] = merged[["score_ratio", "score_set2", "score_set3"]].mean(axis=1)

merged.to_csv("Combined_Final_Score.csv", index=False)

merged
```
